```python

floders=['/media/mrsmile/IA/tesis/data/dataset/asoca/CompressedVersion/Diseased/CTCA', 
         '/media/mrsmile/IA/tesis/data/dataset/asoca/CompressedVersion/Normal/CTCA',]
id=0
data = []
for folder in floders:
    for images in os.listdir(folder):
            root= os.path.join(folder,images)
            idImage, size, spacing, dimension = extraer_info_tecnica(root)
            data.append({
                'Path': root,
                'ImageId' : f"Acosa_{idImage}",
                'Image Size': size,
                'Voxel Size in mm': spacing,
                'Dimension': dimension
            }             
            )
df=pd.DataFrame(data)
```

```python
import os
import pandas as pd


ruta_base_cas = '/media/mrsmile/IA/tesis/data/dataset/imagecas/archive/'

carpetas_cas = [os.path.join(ruta_base_cas, f) for f in os.listdir(ruta_base_cas) 
                if os.path.isdir(os.path.join(ruta_base_cas, f)) and f.startswith('Dataset_')]

data_cas = []

print(f"Procesando {len(carpetas_cas)} bloques de ImageCAS...")

for folder in carpetas_cas:

    for root, dirs, files in os.walk(folder):
        for file in files:
            if file.endswith('.img.nii.gz'):
                path_completo = os.path.join(root, file)
                
                try:
                    idImage, size, spacing, dimension = extraer_info_tecnica(path_completo)
                    data_cas.append({
                        'Path': path_completo,
                        'ImageId': f"Imagecas_{idImage}",
                        'Image Size': size,
                        'Voxel Size in mm': spacing,
                        'Dimension': dimension
                    })
                except Exception as e:
                    print(f"Error procesando {file}: {e}")
df_cas = pd.DataFrame(data_cas)

```

```python
import os, json
import numpy as np

MASKS_DIR = "/media/mrsmile/IA/tesis/data/processed/memmap/masks"
META_DIR  = "/media/mrsmile/IA/tesis/data/processed/memmap/masks_meta"

for f in sorted(os.listdir(META_DIR)):
    name = f.replace(".json", "")
    with open(os.path.join(META_DIR, f)) as jf:
        meta = json.load(jf)
    mmap = np.memmap(os.path.join(MASKS_DIR, f"{name}.dat"),
                     dtype='float32', mode='r', shape=tuple(meta['shape']))
    pct = 100 * mmap.sum() / mmap.size
    flag = " ⚠️" if pct < 0.01 else ""
    print(f"{name}: {pct:.4f}%{flag}")
    del mmap
```

```python
import os

BASE = "/media/mrsmile/IA/tesis/data/dataset/imagecas/archive"
carpetas = [f for f in os.listdir(BASE) if f.startswith('Dataset_')]
print(f"Bloques encontrados: {len(carpetas)}")

# Ver contenido del primer bloque
primer = os.path.join(BASE, sorted(carpetas)[0])
archivos = os.listdir(primer)
print(f"\nContenido de {sorted(carpetas)[0]}:")
for f in sorted(archivos)[:5]:
    print(f"  {f}")

# Contar cuántos tienen label
con_label = 0
sin_label = 0
for carpeta in carpetas:
    ruta = os.path.join(BASE, carpeta)
    archivos = os.listdir(ruta)
    tiene_label = any('label' in f for f in archivos)
    if tiene_label:
        con_label += 1
    else:
        sin_label += 1

print(f"\nCon máscara (.label): {con_label}")
print(f"Sin máscara:          {sin_label}")
```
